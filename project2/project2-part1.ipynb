```python
import json
import pickle
import os
```

```python
def convert_json_to_pkl(json_folder_path, pkl_folder_path):
    # 폴더 내 모든 JSON 파일 목록을 가져오기
    for filename in os.listdir(json_folder_path):
        if filename.endswith(".json"):
            json_file_path = os.path.join(json_folder_path, filename)
            
            # JSON 파일을 읽어들임
            with open(json_file_path, 'r', encoding='utf-8') as json_file:
                data = json.load(json_file)
            
            # pkl 파일로 저장할 경로 설정
            pkl_file_path = os.path.join(pkl_folder_path, f"{os.path.splitext(filename)[0]}.pkl")
            
            # 데이터를 pkl 파일로 저장
            with open(pkl_file_path, 'wb') as pkl_file:
                pickle.dump(data, pkl_file)
            
            print(f"{filename} 변환 완료!")
```

```python
os.path.join('..')
#json_folder_path = r'데이터/Sublabel/SbL/기술_과학/'  # JSON 파일들이 있는 폴더 경로
json_folder_path = r'데이터/Training/02.라벨링데이터/TL_1.발화단위평가_기술_과학/'
pkl_folder_path = r'데이터 저장용'     # pkl 파일들을 저장할 폴더 경로

# 폴더 내 모든 JSON 파일을 pkl로 변환
convert_json_to_pkl(json_folder_path, pkl_folder_path)
```
